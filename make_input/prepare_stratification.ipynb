## Prepare stratification for MITgcm testing

Since stratification will be horizontally homogeneous for now, we just need T and S profiles. 

**Note**: We'll be using a linear equation of state for simplicity, with S and T both active tracers. In file `data`, check that we have the following lines:

`eosType='LINEAR',`

`tAlpha=2.E-4,`

`sBeta=7.4E-4,`
 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
max_depth = 1300
num_levs = 26 # right now we have 26 depth levels, and z-spacing of 50 m, which is very low res.
z = np.linspace(0,max_depth,num_levs) # make depth vector

max_T = 17
min_T = 4
max_S = 36
min_S = 32

def make_linear(topval, botval, zvec):
    slope = (botval-topval)/(zvec[-1]-zvec[0])
    tracer = (slope*zvec) + (topval * np.ones_like(zvec))
    return tracer
```

```python
T = make_linear(max_T, min_T, z)
S = make_linear(min_S, max_S, z)
```

```python
plt.plot(T,z, 'ro')
```

```python
plt.plot(S,z, 'bo')
```
